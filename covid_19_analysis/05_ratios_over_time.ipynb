# 05. Comparative Trends Over Time

This notebook analyzes how different metrics follow each other over time:
1. **Active Cases vs Deaths**
2. **New Deaths vs Recovered**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set(style="whitegrid")
data_dir = "../archive"
```

## 1. Load Data

```python
day_wise = pd.read_csv(os.path.join(data_dir, "day_wise.csv"))
day_wise['Date'] = pd.to_datetime(day_wise['Date'])
day_wise.head()
```

## 2. Comparative Visualizations (Dual Axis)

```python
fig, ax1 = plt.subplots(figsize=(14, 6))

color = 'tab:blue'
ax1.set_xlabel('Date')
ax1.set_ylabel('Active Cases', color=color)
ax1.plot(day_wise['Date'], day_wise['Active'], color=color, linewidth=2, label='Active Cases')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(False)

ax2 = ax1.twinx()

color = 'tab:red'
ax2.set_ylabel('Deaths', color=color)
ax2.plot(day_wise['Date'], day_wise['Deaths'], color=color, linewidth=2, linestyle='--', label='Deaths')
ax2.tick_params(axis='y', labelcolor=color)
ax2.grid(False)

plt.title('Global Active Cases vs Deaths Over Time')
fig.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 6))

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('New Deaths', color=color)
ax1.plot(day_wise['Date'], day_wise['New deaths'], color=color, linewidth=2, label='New Deaths')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(False)

ax2 = ax1.twinx()

color = 'tab:green'
ax2.set_ylabel('Recovered Cases', color=color)
ax2.plot(day_wise['Date'], day_wise['Recovered'], color=color, linewidth=2, linestyle='--', label='Recovered')
ax2.tick_params(axis='y', labelcolor=color)
ax2.grid(False)

plt.title('Global New Deaths vs Recovered Cases Over Time')
fig.tight_layout()
plt.show()
```
